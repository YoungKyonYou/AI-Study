```python
# load general modules
import os
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split, GridSearchCV

import matplotlib.pyplot as plt
import seaborn as sns
```

# 1. Random Forest - <font color='blue'> Classification <font>

```python
# import modules - classification
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix,accuracy_score
from sklearn.datasets import load_breast_cancer
```

### 1) Data preprocessing

```python
#data는 딕셔너리 형태로 되어 있다.
data = load_breast_cancer(return_X_y=False)
```

```python
data.feature_names
```

```python
#two class 문제이다. 양성과 악성 종양에 대한 것이다. 
data.target_names
```

```python
#30개의 input 데이터와 569개의 observation이 있다.
X, y = data.data, data.target
print('{} observations & {} variables'.format(*X.shape))
```

```python
# train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.10,
                                                    random_state=0,
                                                    stratify=y)
```

### 2) Train model

```python
# Grid Search
rf = RandomForestClassifier(random_state=2019)

#n_estimators는 몇 개의 모델을 만들 것인가를 설정한다 (배깅에서 봤었다, 몇개의 데이터를
#부트스트랩해서 학습시킬 것인가 설정) 디폴트 100개이다.
#RF는 모델이 고정되어 있다 DT로 고정되어 있다. 
#DT의 cost function은 GINI와 ENTROPY이다.
#여기서 하려는 것은 최적의 옵션을 찾고 싶은 것이다!!!!! (핵심)
param_grid = {'n_estimators': [10, 50, 100, 200],
              'criterion': ['gini', 'entropy'],
              'max_depth': [None, 2, 4, 6]}
#cross Validation를 한다
#5 cross validation를 한다
rf_cv = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5, scoring='accuracy')
#x train이랑 ytrain으로 피팅을 한다. 기본적으로 5CV하니까 5개씩 만들어지니까 4*2*4=32가 5번만큼
#모델이 만들어지니까 160번 정도 만들어진다 
rf_cv.fit(X_train, y_train)
```

```python
#bset parameter를 봤}더니 아래와 같다. 
print('Best parameters (Grid Search): \n',
      rf_cv.best_params_)
```

```python
#best estimator는 이 모델을 하는데 전체 옵션을 보여주는 것이다.
rf_cv.best_estimator_
#이제 best parameter를 가지고 예측을 해본다
```

### 3) Test and Evaluation

```python
y_pred = rf_cv.best_estimator_.predict(X_test)
#이렇게 써도 best estimator로 predict해준다
#y_pred = rf_cv.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
print(cm)
#밑에 지표에서 2번째 열이 negative 1번째 열이 positive class일 것이다 
#정확도가 높은 것을 볼 수 있다. 
```

```python
acc = accuracy_score(y_test, y_pred)
print('[Accuracy]: {:.2f}%'.format(acc * 100))
```

```python
#f1 score 구하기
from sklearn.metrics import f1_score
```

```python
f1_score(y_test,y_pred)
```

### Visualize confusion matrix

```python
cm = pd.DataFrame(cm, index=data.target_names, columns=data.target_names)

plt.figure(figsize=(6, 5))
sns.heatmap(cm,
            annot=True, annot_kws={'size': 15},
            cmap='binary')
plt.title('[Accuracy]: {:.2f}%'.format(acc * 100))
plt.show()
```

### Feature Importances

```python
# set number of variables to visualize]
#RF에서 중요한게 바로 Feature Importance이다
#변수 중요도이다
#OOB 데이터를 이용해서 각 변수별로 permutation를 했을 때 안 했을 때
#에러 차이를 가지고 변수 중요도를 구했엇다.
#이렇게 변수 중요도를 구할 수 있는 모델이 드물다.
#LR, DT(Information Gain), RF(OOB)를 통해 변수 중요도를 구할 수 있었음
#
topK = 15

# Sort: descending order
#rf_cv.best_estimator까지가 RF이니까 feature importance를 쓸 수 있다.
importances = rf_cv.best_estimator_.feature_importances_
index = np.argsort(importances)[::-1]
```

```python
#이렇게 인덱스가 나온다, 작은 것에서부터 큰것으로까지 나온다.
np.argsort(importances)
```

```python
plt.figure(figsize=(8, 6))
plt.title('Feature Importances')
plt.bar(range(topK),
        importances[index][:topK],
        align='center')
plt.xticks(range(topK), data.feature_names[index][:topK], rotation=90)
plt.xlim([-1, topK])
plt.show()
```

# 2. Random Forest - <font color='blue'> Regression <font>

```python
# import modules - regression
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
```

### 1) Data preprocessing

```python
# load data
data = pd.read_csv(os.path.join('data', 'house regression.csv'))
data.head()
```

```python
print(data.columns)
```

```python
# target variable: price
y = data['price'].values

X = data.drop('price', axis=1)
xcolumns = X.columns.values
X = X.values
```

```python
print('{} observations & {} variables'.format(*X.shape))
```

```python
# train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.20,
                                                    random_state=2019)
```

### 2) Train model

```python
# Grid Search
rf = RandomForestRegressor(random_state=2019)

param_grid = {'n_estimators': [10, 50, 100, 200],
              'max_depth': [None, 2, 4, 6]}
rf_cv = GridSearchCV(estimator=rf, param_grid=param_grid, cv=5)

rf_cv.fit(X_train, y_train)
```

```python
print('Best parameters (Grid Search): \n',
      rf_cv.best_params_)
```

```python
rf_cv.best_estimator_
```

### 3) Test and Evaluation

```python
y_pred = rf_cv.best_estimator_.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print('[Mean Squared Error]: {:.2f}'.format(mse))
print('[R-Square          ]: {:.2f}%'.format(r2 * 100))
```

### Visualize

```python
plt.figure(figsize=(8, 6))
plt.scatter(y_test, y_pred, s=15)
plt.plot(y_test, y_test, alpha=0.5, lw=2, color='red')
plt.xlabel('true', fontsize=12)
plt.ylabel('predict', fontsize=12)
plt.show()
```

```python
idx = y_test.argsort()

plt.figure(figsize=(10, 6))
plt.plot(y_test[idx], color='black', alpha=1.0)
plt.plot(y_pred[idx], color='red', alpha=0.3)
plt.show()
```

### Feature Importances

```python
# set number of variables to visualize
topK = 15

# Sort: descending order
importances = rf_cv.best_estimator_.feature_importances_
index = np.argsort(importances)[::-1]
```

```python
plt.figure(figsize=(8, 6))
plt.title('Feature Importances')
plt.bar(range(topK),
        importances[index][:topK],
        align='center')
plt.xticks(range(topK), xcolumns[index][:topK], rotation=90)
plt.xlim([-1, topK])
plt.show()
```
